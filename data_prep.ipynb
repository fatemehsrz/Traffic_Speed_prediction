```python

import pandas as pd

df_adj = pd.read_csv('./data/los_adj.csv')

df_adj.head()
```

```python
f_adj= open( './data/los_edgelist.txt', 'w')

vis_los= open( './data/los_vis_edgelist.txt', 'w')

nodes_los=set()

edges_los=[]

for index, row in df_adj.iterrows():
    
    for j in range(len(row)):
        
        if row[j]>0:
    
          #print(index, j, row[j] 
            
          nodes_los.add(int(index))
          nodes_los.add(int(j))
            
          edges_los.append( [int(index), int(j) ])  
            
          f_adj.write(str(index)+" "+ str(j)+" "+str(row[j]))
          f_adj.write('\n')
          vis_los.write(str(index)+" "+ str(j))
          vis_los.write('\n')
        
        
f_adj.close()
vis_los.close()

print(len(edges_los))
```

```python
import numpy as np
import random



df_los = pd.read_csv('./data/los_speed.csv')

features_los=[]
los_speed=[]
los_sensor_speed=[]

random_num=random.randint(0,20)

for index, row in df_los.iterrows():
    
    row_fill=[]
    
    los_sensor_speed.append(row[random_num])
    
    for j in row:
        
       row_fill.append(j)
       los_speed.append(j)
    
    features_los.append(row_fill)
        
   
print(len(features_los))    
      
print(len(features_los[0]) ) 

print(min(los_speed))
print(max(los_speed))
print(np.mean(los_speed))

print(len(los_sensor_speed))
```

```python
import numpy as np
import matplotlib.pyplot as plt
    
from matplotlib.pyplot import figure   
    
figure(figsize=(18, 5), dpi=80)  

    
    
Data = {'h': list(range(len(los_sensor_speed))),
        's': los_sensor_speed
       }

  
dff = pd.DataFrame(Data,columns=['h','s'])
  
plt.plot(dff['h'], dff['s'], color='r' ) #marker='o'
plt.title('Speed reported by a random sensor in Los Angeles', fontsize=20)
plt.xlabel('Min', fontsize=20)
plt.ylabel('Speed', fontsize=20)

plt.tick_params(axis='both',labelsize=20)

plt.grid(True)
#plt.show()

plt.savefig('los_sensor2.pdf', bbox_inches='tight')
```

```python
from scipy import stats
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
fig.set_size_inches(9, 6)


ax.grid(True)
ax.grid(linestyle=':')
sz_speed.sort()
hmean = np.mean(sz_speed)
hstd = np.std(sz_speed)
pdf = stats.norm.pdf(sz_speed, hmean, hstd)
plt.plot(sz_speed, pdf, color= 'blue', linewidth=3)

plt.xlabel('Speed (mph)', fontsize=25)
plt.ylabel('Frequency', fontsize=25)
plt.title('Speed distribution in Shenzen', fontsize=23)


plt.tick_params(axis='both',labelsize=25)
fig.tight_layout()

#plt.show() 

plt.savefig('sz_speed.pdf', bbox_inches='tight')
```

```python
import pandas
from collections import defaultdict
import csv

results_holder_r=defaultdict(list)


with open('./data/results_los.csv', 'r') as f:
    
    
    
    reader = csv.reader(f)
    next(reader, None)
    
    for row in reader:
       print(row[0], row[2] ,row[3])
        
       results_holder_r[row[0]].append(float(row[2]))
       results_holder_r[row[0]].append(float(row[3]))
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import numpy as np
#fig, ax = plt.subplots()



figure(figsize=(19, 6), dpi=80)



height=[]

for i in results_holder_r:
    
    height.append(results_holder_r[i][0])
    

bars = tuple(results_holder_r)
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, width = 0.3, color='r')
plt.xticks(y_pos, bars)


plt.title('Error reported for car speed prediction', fontsize=19)

plt.ylabel('RMSE', fontsize=19)
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)

plt.tick_params(axis='both',labelsize=15)
fig.tight_layout()

plt.savefig('comparison_los1.pdf', bbox_inches='tight')
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import numpy as np


figure(figsize=(22, 7), dpi=80)



height=[]

for i in results_holder_r:
    
    height.append(results_holder_r[i][1])
    

bars = tuple(results_holder_r)
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, width = 0.3, color='r')
plt.xticks(y_pos, bars)


plt.title('Error reported for car speed prediction', fontsize=20)

plt.ylabel('MAE', fontsize=19)
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)

plt.tick_params(axis='both',labelsize=17)
fig.tight_layout()

plt.savefig('comparison_los2.pdf', bbox_inches='tight')
```
